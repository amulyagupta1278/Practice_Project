```python
class DeepLearningStudent():
    def __init__(self,name,age): #dunder init
        self.name = name
        self.age = age
        self.course = 'CSF425 Deep Learning'
        print('got a student of age', self.age)
    
    def twice_the_age(self):
        return self.age*2
    
    def four_times_the_age(self):
        twice = self.twice_the_age()
        return twice*2

palaash = DeepLearningStudent('Palaash',18)
palaash.twice_the_age()
```

## Inheritance

```python
class BITS_student():
    def __init__(self,name,age, is_PS = True, campus ='Pilani'):
        self.name = name
        self.age = age
        self.is_PS = is_PS
        self.campus = campus
        print('This is a student of BITS Pilani ',campus, 'campus of batch', 2021 -self.age + 18)
    
    def print_id(self):
        # 2018AxPSxxxxP
        ps = 'PS' if self.is_PS else 'TS'
        return str(2021 - self.age + 18) + 'Ax' + ps + 'xxxx' +  self.campus[0].upper() 

palaash = BITS_student('Palaash',21) 
```

```python
palaash.print_id()
```

```python
class DeepLearningStudent(BITS_student):
    pass
    def __init__(self,name,age): #dunder init
        super().__init__(name,age)
        self.name = name
        self.age = age
        self.course = 'CSF425 Deep Learning'
        print('got a student of age', self.age)
    
    def twice_the_age(self):
        return self.age*2
    
    # def four_times_the_age(self):
    #     twice = self.twice_the_age()
    #     return twice*2

palaash = DeepLearningStudent('Palaash',18)
# palaash.twice_the_age()
```
